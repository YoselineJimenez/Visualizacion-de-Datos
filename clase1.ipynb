```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
events =sb.events(match_id=3857289)
sns.heatmap(events.isna(),cmap="Reds")

```

```python
events
```

```python
columnas=['location','minute','pass_end_location','player,'pass_recipient','period', 'team','type']
df=events[columnas]
df['type'].unique()
```

```python
passes =df[df['type']]=='Pass'
passes.drop(columns='type',inplace= True)
passes.index=range(len(passes))
passes.head()
```

```python
passes=pd.concat([passes, passes.location.apply(pd.Series)], axis=1)
passes=pd.concat([passes,passes.pass_end_location.apply(pd.Series)])
```

```python
from mplsoccer import Pitch
# https://mplsoccer.readthedocs.io/en/latest/installation.html
fig, axs =plt.subplots()
sns.scatterplot(data=passes, x='X0', y='Y0',ax=axs,hue='team')
axs.set_title("Localizacion de pases amExicp")
plt.show()
```
