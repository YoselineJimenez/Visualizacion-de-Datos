```python
import pandas as pd
import folium 
from folium.plugins import MarkerCluster
```

```python
base_hoteles=pd.read_csv("Base_hoteles.csv")
```

```python
base_hoteles.head(5)
```

```python
#Aqui creamos un mapa centrado en una coordenada inicial
mapa = folium.Map(location=[20, 0], zoom_start=2)
```

```python

# Definir una función para determinar el color según el score

def color(score):
    if score == 10:
        return 'darkgreen'
    elif score == 9:
        return 'green'
    elif score == 8:
        return 'orange'
    elif score ==7:
        return 'red'
    elif score ==6:
        return 'blue'
    elif score ==5:
        return 'purple'
    else:
        return 'darkred'
```

```python
marker_cluster = MarkerCluster().add_to(mapa)
```

```python
for index, row in base_hoteles.iterrows():
    folium.Marker(
        location=[row['Latitud'], row['Longitud']],
        popup=f"Pais: {row['Pais']}<br>Score: {row['Score']}",
        tooltip=row['Nombre'],
        icon=folium.Icon(color=color(row['Score']))
    ).add_to(marker_cluster)

```
