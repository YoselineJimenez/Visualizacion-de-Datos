```python
# https://data.london.gov.uk/dataset/eu-referendum-results

url_data="https://data.london.gov.uk/download/eu-referendum-results/52dccf67-a2ab-4f43-a6ba-894aaeef169e/EU-referendum-result-data.csv"
```

```python
df=pd.read_csv(url_data)
df.head()
```

```python
df.columns
```

```python
columnas=['Region', 'Area', 'Electorate','ExpectedBallots',
         'VerifiedBallotPapers', 'Pct_Turnout', 'Remain',
         'Leave', 'Rejected_Ballots','Pct_Remain', 'Pct_Leave']
df=df.loc[:,columnas]
```

```python
df.head()
```

```python
df.VerifiedBallotPapers/df.Electorate
```

```python
# Mapa coroplético (https://geopandas.org/en/stable/)
!pip install geopandas
import geopandas as gpd
```

```python
# Geopandas utiliza archivos shp & json
url_uk="https://raw.githubusercontent.com/martinjc/UK-GeoJSON/master/json/administrative/gb/topo_lad.json"
```

```python
# Los geoDF siempre deben de tener una columna geometry
uk = gpd.read_file(url_uk)
uk = uk.loc[:, ["LAD13NM", "geometry"]]
uk.columns=["Area", "geometry"]
```

```python
uk.shape, df.shape
```

```python
df=df[df["Area"].isin(uk["Area"].values)]
df.shape
```

```python
# Para implementar un join, se cambia el índice de ambos
# df, y se ordenan previo al join
uk.set_index("Area", inplace=True)
df.set_index("Area", inplace=True)
uk.sort_index(inplace=True)
df.sort_index(inplace=True)
```

```python
final=uk.join(df)
final.head()
```

```python
type(final)
```

```python
dir(final.geometry)
```

```python
Cities=["City of London", "Glasgow City", "City of Edinburgh", "Manchester", "Liverpool"]
fig, axs = plt.subplots()
sns.scatterplot(data=final, x="density",
                y="Pct_Remain", label="Remain")
for c in Cities:
  sns.scatterplot(data=final[final.Area==c], x="density",
                  y="Pct_Remain", label=c)
axs.set_xscale("log")
#axs.set_yscale("log")
axs.set_ylabel("% Voters")
axs.legend(bbox_to_anchor =(1,1))
axs.grid()
plt.show()
```

```python
final.plot(column="Region", legend=True,
           legend_kwds={"loc": "center left",
                        "bbox_to_anchor": (1, 0.5)})
plt.title("Regiones geográficas de UK")
plt.axis("off")
plt.show()
```

```python
from matplotlib.colors import ListedColormap
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b',
          '#e377c2', '#7f7f7f', '#bcbd22', '#17becf', '#aec7e8']
paleta=ListedColormap(colors)

final.plot(column="Region", legend=True, legend_kwds={"loc": "center left", "bbox_to_anchor": (1, 0.5)}, cmap=paleta)
plt.title("Regiones geográficas de UK")
plt.axis("off")
plt.show()
```

```python
titulos=["Regiones geográficas de UK",
         "Número de votantes"]
cols=["Region", "Electorate"]
fig, axs = plt.subplots(1,2, figsize=(10,5))
for n in range(2):
  if n==0:
    final.plot(column=cols[n], ax=axs[n], cmap=paleta)
  else:
    final.plot(column=cols[n], legend=True,
              legend_kwds={"shrink":0.5},
              ax=axs[n])
  axs[n].set_title(titulos[n])
  axs[n].axis("off")
plt.show()
```

# Diagrama Sankey

```python
# https://www.w3schools.com/colors/colors_hsl.asp
# hsl(hue, saturation, lightness)
# hue -> 0 a 360°
# sat -> 0 a 100%
# lig -> 0 a 100%
negro="hsl(0, 0%, 0%)"
blanco="hsl(0, 0%, 100%)"
```

<img src="https://thenewcode.com/assets/images/hsl-color-wheel.png" width=500>

```python
rojo="hsl(0, 100%, 50%)"
verde="hsl(120, 100%, 50%)"
azul="hsl(240, 100%, 50%)"

amarillo="hsl(60, 100%, 50%)"
cyan="hsl(180, 100%, 50%)"
magenta="hsl(300, 100%, 50%)"

azul_claro="hsl(240, 100%, 75%)"
rojo_claro="hsl(0, 100%, 75%)"
gris="hsl(0, 0%, 25%)"
```

```python
# https://plotly.com/python/sankey-diagram/
long_uk.head(3)
```

```python
long_uk.Region.unique()
```

```python
loc_uk=long_uk[long_uk.Region=='North West']
loc_uk
```

```python
# Seleccionando los nodos
node_names=np.concatenate([loc_uk.Area.unique(), loc_uk.variable.unique()])
# Creando un diccionario {node:index}
nodes={n:i for i, n in enumerate(node_names)}
nodes
```

```python
nodes_df= loc_uk.iloc[:,1:].applymap(lambda cell: nodes[cell] if
                           cell in nodes.keys() else cell)
```

```python
node = dict(
      pad = 15,
      thickness = 20,
      line = dict(color = "black", width = 0.5),
      label = node_names,
      color = "blue"
    )
```

```python
nodes_df.columns
```

```python
link = dict(
      source = nodes_df["Area"],
      target = nodes_df["variable"],
      value = nodes_df["value"]
  )
```

```python
data_sankey=go.Sankey(node = node,
                      link = link)
fig = go.Figure(data=data_sankey)

fig.update_layout(title_text="Mi primer diagrama de Sankey",
                  font_size=10)
fig.show()
```

```python
loc_uk2=long_uk[(long_uk.Region=='North West') &
                (long_uk.Area.isin(["Manchester", "Liverpool", "Bolton",
                                    "St. Helens", "Blackpool", "Burnley"]))]
node_names2=np.concatenate([loc_uk2.Area.unique(), loc_uk2.variable.unique()])
nodes2={n:i for i, n in enumerate(node_names2)}
nodes_df2= loc_uk2.iloc[:,1:].applymap(lambda cell: nodes2[cell] if
                           cell in nodes2.keys() else cell)
```

```python
conexiones=[]
for x in nodes_df2.variable:
  if x==6:
    conexiones.append(azul_claro)
  if x==7:
    conexiones.append(rojo_claro)
  if x==8:
    conexiones.append(gris)
```

```python
node = dict(pad = 15,
      thickness = 20,
      line = dict(color = "black", width = 0.5),
      label = node_names2, # El orden de las etiquetas marcará el orden de los colores
      color = [negro, cyan, verde, rojo, rojo_claro, amarillo, azul, rojo, negro])
link = dict(source = nodes_df2["Area"],
      target = nodes_df2["variable"],
      value = nodes_df2["value"],
      color=conexiones) # Aquí el orden de colores depende del dataframe usado en link
```

```python
data_sankey=go.Sankey(node = node,
                      link = link)
fig = go.Figure(data=data_sankey)

fig.update_layout(title_text="Diagrama de Sankey, Brexit 2016",
                  font_size=10)
fig.show()
```
