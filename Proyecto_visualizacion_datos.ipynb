```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
```

```python
#creamos el grafico 
fig, ax = plt.subplots(figsize=(10, 6))

##Aqui asignamos los colores 
colormap = plt.cm.rainbow  
colors = [colormap(i) for i in np.linspace(0, 1, len(df['Pais']))]


ax.axhline(0, color='black', linewidth=0.5)  # este es el  Eje x
ax.axvline(0, color='black', linewidth=0.5)  # este es el Eje y

for i in range(len(df['Pais'])):
    ax.scatter(df['x'].iloc[i], df['y'].iloc[i], 
               s=df['x'].iloc[i] * 100,  # Escala adecuada para que el tamaño de la burbuja represente x
               color=colors[i], alpha=0.6, label=df['Pais'].iloc[i])


for i in range(len(df['Pais'])):
    plt.text(df['x'].iloc[i], df['y'].iloc[i], f"{df['x'].iloc[i]}%", 
             ha='center', va='center', color='black', fontsize=8)

# Ajustar leyenda
ax.legend(title="País", loc="lower left", bbox_to_anchor=(1, 0))

# Título y etiquetas de los ejes
plt.title('Proporción del Valor del Mercado de Drogas Mundial vs PIB por País')

# Mostrar el gráfico
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
##Aqui creamos una estructura de datos en forma de un diccionario de listas
data = {
    "Año": [2020, 2021, 2022, 2020, 2021, 2022, 2020, 2021, 2022],
    "Pais": ["Mexico"] * 9,
    "Rubro": ["Cannabis", "Cannabis", "Cannabis", "Tabaco", "Tabaco", "Tabaco", "Alcohol", "Alcohol", "Alcohol"],
    "impuestos_MD": [216,199,171,244,254,266,96,111,133]
}

df = pd.DataFrame(data)

# Creamos un grafo en linea 
plt.figure(figsize=(12, 8))

# Dibujamos en líneas para cada rubro y agregamos las etiquetas 
for rubro in df['Rubro'].unique():
    df_rubro = df[df['Rubro'] == rubro]
    plt.plot(df_rubro['Año'], df_rubro['impuestos_MD'], marker='o', label=rubro)
    for i in range(len(df_rubro)):
        plt.text(df_rubro['Año'].iloc[i], df_rubro['impuestos_MD'].iloc[i], f"{df_rubro['impuestos_MD'].iloc[i]:.2f}", ha='right')

# Aqui Añadimos las etiquetas y título 
plt.xlabel('Año')
plt.ylabel('Impuestos (MD)')
plt.title('Comparación de Ingresos Fiscales de Cannabis, Tabaco y Alcohol en México (2020-2022)')
plt.legend(title='Rubro')
plt.grid(True)

# Configuaramos  los ticks del eje X para mostrar solo los años completos
plt.xticks([2020, 2021, 2022])

# Mostrar el gráfico
plt.show()
```

```python

import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
import numpy as np

# Creamosun  DataFrame con los datos de recaudacuion fiscal
data = {
    "Paises": ["Egypt", "United States", "England", "Australia", "Russia", "Germany", "Mexico", "Canada"],
    "Impuesto Tabaco": [0.73, 0.42, 0.82, 0.56, 0.47, 0.72, 0.67, 0.69],
    "mdd": [384.87, 354.48, 237.35, 138.36, 128.9, 114.7, 171.05, 124]
}

df = pd.DataFrame(data)

# Obtener datos geográficos
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Ajustar nombres de países para el merge
df['Paises'] = df['Paises'].replace({
    "United States": "United States of America",
    "England": "United Kingdom"
})

# Merge con DataFrame para tener coordenadas geográficas
df_geo = world.merge(df, how="left", left_on="name", right_on="Paises")

# Aqui Creampos el gráfico de mapa
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
world.boundary.plot(ax=ax)

# Dibujamos unas burbujas en el mapa
divider = make_axes_locatable(ax)
ax_table = divider.append_axes("right", size="20%", pad=0.1)

colors = plt.cm.rainbow(np.linspace(0, 1, len(df_geo)))

for i, row in df_geo.iterrows():
    if pd.notna(row['mdd']):
        ax.scatter(row.geometry.centroid.x, row.geometry.centroid.y,
                   s=row['mdd'], alpha=0.6, color=colors[i])
        ax.text(row.geometry.centroid.x, row.geometry.centroid.y,
                f"{row['mdd']:.2f}", fontsize=8, ha='center', color='black')

# Creamos una tabla con etiquetas 
table_data = df[['Paises', 'Impuesto Tabaco']].values
ax_table.axis('off')
table = ax_table.table(cellText=table_data, colLabels=['Paises', 'Impuesto Tabaco'], cellLoc='center', loc='center')

# Ajustamos las fuente de las tables 
table.auto_set_font_size(False)
table.set_fontsize(8)
table.scale(1.2, 1.2)

# Título y ajustes finales
ax.set_title('Posibles Impuestos al Cannabis de acuerdo al impuesto por Tabaco', fontsize=16)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos de equivalencias que tenemos 
data = {
    'Equivalencia': [
        'Construcción de 1 escuela', 
        'Construcción de 1 hospital', 
        'Mantenimiento de carreteras (1 año)', 
        'Financiación de programas sociales (1 año)'
    ],
    'Costo (millones de USD)': [20, 50, 30, 10],
    'Cantidad': [
        171 / 20, 
        171 / 50, 
        171 / 30, 
        171 / 10
    ]
}

# Creamos aca  el DataFrame
df = pd.DataFrame(data)

# Creamos aca el grafico de barras 
plt.figure(figsize=(10, 6))
plt.barh(df['Equivalencia'], df['Cantidad'], color=['blue', 'green', 'orange', 'purple'])

# Añadimos  etiquetas y título
for index, value in enumerate(df['Cantidad']):
    plt.text(value, index, f'{value:.1f}', va='center')
    
plt.xlabel('Cantidad')
plt.title('Equivalencias de la Recaudación Anual de 171 Millones de Pesso')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos proporcionados
data = {
    "Año": [2017, 2018, 2019, 2020, 2021, 2022],
    "Area": [26500, 61700, 32700, 77300, 107059, 0],
    "Burbuja": [122800, 149300, 211000, 243700, 321000, 428059],
    "Leyenda_Area": ["+26,500", "+61,700", "+32,,700", "+77,300", "+107,059", ""]
}

# Crear la figura y el eje
fig, ax = plt.subplots(figsize=(12, 8))  # Aumentamos la altura de la figura

# Gráfico de área para raea y años 
ax.fill_between(data["Año"], data["Area"], color="#FFCC00", alpha=0.6)

# Añadir las etiquetas dentro del área
for i in range(len(data["Año"])):

    ax.text(data["Año"][i], data["Area"][i] / 2, data["Leyenda_Area"][i], fontsize=12, ha='center', va='center', color='black')

# Gráfico de burbujas con tamaños ajustados según el valor lo dividiremos para ajustar a la grafica 
for i in range(len(data["Año"])):
    bubble_size = data["Burbuja"][i] / 1000  # Ajustar el factor de escala según sea necesario
    ax.scatter(data["Año"][i], data["Burbuja"][i], s=bubble_size, color="#006600", alpha=0.6)
    # Aqui Convertimos valores a miles y añadir 'K'
    burbuja_k = f"{data['Burbuja'][i] / 1000:.1f}K"
    ax.text(data["Año"][i], data["Burbuja"][i], burbuja_k, fontsize=10, ha='center', va='bottom', color='black')

# Títulos y etiquetas
ax.set_title("Crecimiento del empleo en la industria del cannabis en los últimos 5 años", fontsize=14)
ax.set_xlabel("Año", fontsize=12)
ax.set_ylabel("Empleos", fontsize=12)
ax.grid(True, linestyle='--', alpha=0.5)

# Configurar los ticks del eje x y el rango del eje y
plt.xticks(data["Año"])
plt.yticks(range(0, 500000, 50000))

# Mostrar gráfico
plt.tight_layout()
plt.show()
```
